# Import the libraries

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Load the dataset

```python

df=pd.read_csv(r"D:\EPL_20_21.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
#to summarize the statistics about dataset
df.describe()
```

```python
#check the total number of null values
df.isnull().sum()
```

```python
#create two new columns
df['MinsPerMatch']=(df['Mins']/df['Matches']).astype(int)
df['GoalsPerMatch']=(df['Goals']/df['Matches']).astype(float)
df.head()
```

```python
#total goals
Total_Goals=df['Goals'].sum()
Total_Goals
```

```python
#total penalty goals
Total_PenaltyGoals=df['Penalty_Goals'].sum()
Total_PenaltyGoals
```

```python
#Penalty attempts
TotalPenaltyattempts=df['Penalty_Attempted'].sum()
TotalPenaltyattempts
```

```python
##pie chart for penalties missed vs scored
plt.figure(figsize=(13,6))
pl_not_scored=df['Penalty_Attempted'].sum()-Total_PenaltyGoals
data=[pl_not_scored,Total_PenaltyGoals]
labels=['Penalties missed','Penalties Scored']
color=sns.color_palette("Paired")
plt.pie(data,labels=labels,colors=color,autopct='#ff214c')
plt.show
```

```python
##unique Position
df['Position'].unique()
```

```python
#Total FW players
df[df['Position']=='FW']
```

```python
##Players from different nation
np.size((df['Nationality'].unique()))
```

```python
##most players from which countries
nationality=df.groupby('Nationality').size().sort_values(ascending=False)
nationality.head(10).plot(kind='bar',figsize=(12,6),color=sns.color_palette("magma"))
```

```python
##club with maximum players in their squad

df['Club'].value_counts().nlargest(5).plot(kind="bar",color=sns.color_palette('viridis'))
```

```python
#club with least players in their squad

df['Club'].value_counts().nsmallest(5).plot(kind="bar",color=sns.color_palette('viridis'))
```

```python
#players based on age groups
Under20=df[df['Age']<=20]
Under20_25=df[(df['Age']>20) & (df['Age']<=25)]
Under25_30=df[(df['Age']>25) & (df['Age']<=30)]
Above30=df[df['Age']>30]
```

```python
x=np.array([Under20['Name'].count(),Under20_25['Name'].count(),Under25_30['Name'].count(),Above30['Name'].count()])
mylabels=["<=20",">20 & <=25",">25 & <=30",">30"]
plt.title("Total Players with age",fontsize= 20)
plt.pie(x,labels=mylabels,autopct="%.1f%%")
plt.show
```

```python
#total under 20 players in each clubs
players_under_20=df[df['Age']<20]
players_under_20['Club'].value_counts().plot(kind="bar",color=sns.color_palette("cubehelix"))
```

```python
#under 20 players in manU
players_under_20[players_under_20["Club"]=='Manchester United']
```

```python
#Under 20 Players in Chelsea
players_under_20[players_under_20["Club"]=='Chelsea']
```

```python
##Avg  age of players in each club
plt.figure(figsize=(12,6))
sns.boxplot(x='Club',y='Age',data=df)
plt.xticks(rotation=90)
```

```python
num_player=df.groupby('Club').size()
data=(df.groupby('Club')['Age'].sum())/num_player
data.sort_values(ascending=False)
```

```python
##total assists by each clubs

Assists_by_clubs=pd.DataFrame(df.groupby('Club',as_index=False)['Assists'].sum())
sns.set_theme(style='whitegrid',color_codes=True)
ax=sns.barplot(x='Club',y='Assists',data=Assists_by_clubs.sort_values(by='Assists'),palette='Paired')
ax.set_xlabel('Club',fontsize=30)
ax.set_ylabel('Assists',fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]=(20,8)
plt.title('Plot of Clubs vs Total Assists',fontsize=20)
```

```python
#top 10 Assists
top_10_Assists=df[['Name','Club','Assists','Matches']].nlargest(n=10,columns='Assists')
top_10_Assists
```

```python
Goals_by_clubs=pd.DataFrame(df.groupby('Club',as_index=False)['Goals'].sum())
sns.set_theme(style='whitegrid',color_codes=True)
ax=sns.barplot(x='Club',y='Goals',data=Goals_by_clubs.sort_values(by='Goals'),palette='Paired')
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Goals",fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]=(20,8)
plt.title('Plot of Clubs vs Total Goals',fontsize=20)
```

```python
##most Goals by Players
top_10_goals=df[['Name','Club','Goals','Matches']].nlargest(n=10,columns='Goals')
top_10_goals
```

```python
#pie chart-goals with assists and without assists
plt.figure(figsize=(14,7))
assists=df['Assists'].sum()
data=[Total_Goals-assists,assists]
labels=['Goals without assists','Goals with assists']
color=sns.color_palette("Set1")
plt.pie(data,labels=labels,colors=color,autopct= '%.0f%%')
plt.show()
```

```python
#top 10 players with most yellow cards
epl_yellow=df.sort_values(by='Yellow_Cards',ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.title('Players with the most yellow Cards')
C=sns.barplot(x=epl_yellow['Name'],y=epl_yellow['Yellow_Cards'],label='Players',color='yellow')
plt.ylabel('Number of yellow Cards')
C.set_xticklabels(C.get_xticklabels(),rotation=45)
C
```

```python
#top 10 players with most red cards
epl_red=df.sort_values(by='Red_Cards',ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.title('Players with the most Red Cards')
D=sns.barplot(x=epl_red['Name'],y=epl_red['Red_Cards'],label='Players',color='red')
plt.ylabel('Number of red Cards')
D.set_xticklabels(C.get_xticklabels(),rotation=45)
D
```
